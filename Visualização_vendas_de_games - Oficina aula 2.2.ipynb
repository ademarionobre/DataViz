# Visualiação de vendas de games

Prática utilizando gráficos de pizza e gráficos de barra para visualizar proporções entre categorias.
Dados de vendas de videogames estão disponíveis no link: https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro04/oficina/arquivos/dadosoficina04.txt

Questões a serem respondidas
- 1. Criar uma visualização que mostre a proporção de vendas global entre gêneros de jogos da empresa Nintendo.
- 2. Criar uma visualização que mostre a proporção de quantidade de jogos por plataforma a partir do ano de 2015.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
games = pd.read_csv('https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro04/oficina/arquivos/dadosoficina04.txt')
games
```

```python
games.describe()
```

```python
empresa = games.groupby(by='Publisher').count().sort_values('Global_Sales', ascending=False)
empresa
```

```python
plataforma = games[games['Year']>=2015.0]
plataforma = plataforma.groupby(by='Platform').count().sort_values('Name', ascending=False)
plataforma
```

```python
nintendo = games.loc[games['Publisher']=='Nintendo']
nintendo = games.groupby(by='Genre').count().sort_values('Global_Sales', ascending=False)
nintendo
```

```python
generos = games.groupby(by='Genre').count().sort_values('Global_Sales', ascending=False)
generos
```

________________________

1. Visualizando gráficos vendas globais entre gêneros de jogos da empresa Nintendo.

```python
porcentagens = nintendo['Global_Sales']*100/nintendo['Global_Sales'].sum()
plt.style.use('bmh')
plt.figure(figsize=(15,5))
plt.bar(nintendo.index, porcentagens)
plt.title('Proporção de vendas globais de jogos por gênero da Nintendo')
plt.show()
```

```python
plt.pie(x=nintendo['Global_Sales'],
        labels = generos.index,
        autopct='%1.1f%%',
        startangle=200,
        pctdistance=0.80,
        radius=2.5)
plt.title('Proporção de vendas globais de jogos por gênero da Nintendo')
plt.show()
```

2. Visualização da proporção de quantidade de jogos por plataforma a partir do ano de 2015.

```python
porcentagens = plataforma['Name']*100/plataforma['Name'].sum()
plt.style.use('seaborn')
plt.figure(figsize=(15,5))
plt.bar(plataforma.index, porcentagens)
plt.title('Proporção de de jogos por plataforma de 2015 em diante')
plt.show()
```

```python
plt.pie(x=plataforma['Name'],
        labels=plataforma.index,
        autopct='%1.1f%%',
        startangle=90,
        pctdistance=0.90, radius=2)
my_circle=plt.Circle((0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.title('Proporção de de jogos por plataforma de 2015 em diante')
plt.show()
```

Ambas visualizações de gráfico pizza/rosca apresentaram valores de rótulos o que auzilia na identificação da representação
das categorias o que não ocorre nos gráficos de barras.
Os gráficos de barras apresentam os dados com melhor percepção da distribuição no visual.
