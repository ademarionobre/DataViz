# Oficina 4.2. Visualizando distribuição conjunda de dados

Nesta oficina você irá analisar distribuições conjuntas de um conjunto de dados sobre carros. Esse conjunto de dados contém informações sobre: consumo de galão de combustível por milha (miles per galon, mpg), número de cilindros, cilindrada, cavalos de potência, peso, aceleração, ano do modelo, país de origem e nome do carro. Esse conjunto de dados chama-se “mpg” e é bastante famoso em estatística, sendo disponibilizado diretamente pela biblioteca seaborn sem a necessidade de baixar os dados para a sua máquina.

```python
# importando bibliotecas e estilo de visualização.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
# carregar dados direto do seaborn.
mpg = sns.load_dataset("mpg")
mpg.head()
```

```python
mpg.describe()
```

##### Criar uma visualização que mostre a distribuição conjunta de todas as variáveis quantitativas, diferenciando pelo país de origem (‘origin’).  
1. Quais variáveis possuem uma maior correlação?  
2. É possível distinguir o país de origem a partir da distribuição de uma variável?
3. E de uma distribuição conjunta?

```python
sns.pairplot(mpg, hue='origin')
plt.title('Distribuições conjuntas de todas as variáveis quantitativas pelo país', x=-3,
         pad=1030, color='blue');
```

1. Quais variáveis possuem uma maior correlação?
 - O parâmetro mpg(consumo) possui correlação com displacemente(deslocamento), horsepower(potência) e weight(peso) dos carros. Além do horsepower, weigth e acceleration estarem correlacionados entre eles. Destacando que o comportamento dos EUA possui maior distribuição dos dados.  
 
 
2. É possível distinguir o país de origem a partir da distribuição de uma variável?
 - No gráfico de densidade é possível visualizar pela cor o país gerando uma separação em cada varíavel observável.  
 
 
3. E de uma distribuição conjunta?
 - Não é possível. Ajuda ter uma noção geral do comportamento dos países, mesmo as cores azul (USA), laranja (Japão) e verde (Europa) estarem definidas, mas a medida que os dados aumentam fica difícil distinguir pela sobreposição de cores.

```python
sns.pairplot(mpg, kind='reg');
plt.title('Distribuições conjuntas de todas as variáveis com retas de regressão', x=-3,
         pad=1030, color='darkblue');
```

- Observando os mesmos dados com a reta de regressão.

##### Escolha de uma distribuição conjunta.
##### Visualização da distribuição conjunta com reta de regressão e as ditribuições marginais.

```python
sns.jointplot(x='weight', y='mpg', kind='reg', data=mpg)
plt.title('Distribuição conjunta de correlação do consumo de combustível e peso do carro', x=-1,
         pad=80, color='darkblue');
```
